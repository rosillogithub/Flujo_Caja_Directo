Después agregar información para cada registro a las columnas "grupo" y "nombre_grupo" de la siguiente forma:
1. Acceder a la base de datos MySQL llamada "flujo_caja_directo" usando el contenido del archivo config.ini
2. Buscar el contenido del campo "CUENTA" del dataframe "df_Mov_98" en la columna "cod_cont" de la tabla "tbl_fcdir_contab" y guardar el contenido de la columna "id_grp"
3. Buscar el contenido del campo "grupo" del dataframe "df_Mov_98" en la columna "id_grp" de la tabla "tbl_fcdir_grupos" y guardar el contenido de la columna "name" en la columna "nombre_grupo" del dataframe "df_Mov_98"

```python
import mysql.connector
import pandas as pd
import configparser

# Cargar el archivo config.ini para obtener los detalles de la base de datos
config = configparser.ConfigParser()
config.read('config.ini')

# Obtener los detalles de la base de datos desde el archivo config.ini
db_config = config['database']

# Establecer la conexión a la base de datos MySQL
try:
    connection = mysql.connector.connect(
        user=db_config['user'],
        password=db_config['password'],
        host=db_config['host'],
        database=db_config['database']
    )

    if connection.is_connected():
        cursor = connection.cursor()

        # Iterar sobre cada fila del DataFrame y buscar el valor de CUENTA en la tabla tbl_fcdir_contab
        for index, row in df_Mov_98.iterrows():
            cuenta = row['CUENTA']
            query_contab = f"SELECT id_grp FROM tbl_fcdir_contab WHERE cod_cont = {cuenta}"
            cursor.execute(query_contab)
            result_contab = cursor.fetchone()

            # Si se encuentra un resultado, asignar el valor de id_grp a la columna 'grupo'
            if result_contab:
                grupo = result_contab[0]
                df_Mov_98.at[index, 'GRUPO'] = grupo

                # Buscar el nombre del grupo en la tabla tbl_fcdir_grupos
                query_grupo = f"SELECT name FROM tbl_fcdir_grupos WHERE id_grp = {grupo}"
                cursor.execute(query_grupo)
                result_grupo = cursor.fetchone()

                # Si se encuentra un resultado, asignar el nombre del grupo a la columna 'nombre_grupo'
                if result_grupo:
                    nombre_grupo = result_grupo[0]
                    df_Mov_98.at[index, 'NOMRE_GRUPO'] = nombre_grupo

        # Cerrar el cursor y la conexión
        cursor.close()
        connection.close()

except mysql.connector.Error as error:
    print("Error al conectarse a la base de datos:", error)

# Mostrar el DataFrame actualizado
df_Mov_98.head()

```

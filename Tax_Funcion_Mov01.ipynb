```python
import pandas as pd

# Define cod_cont_fcd
cod_cont_fcd = [110505, 112005]

def filter_group(group):
    if group['CUENTA'].isin(cod_cont_fcd).any():
        return group

# Group df_Mov_99 by DOCUMENTO and apply filter_group to each group
df_mov_fcd = df_Mov_99.groupby('DOCUMENTO').apply(filter_group)

df_mov_fcd.head(10)
```
